#**LeetCode: Container With Most Water**

# **Problem**


**Reference:**

https://leetcode.com/problems/container-with-most-water/description/

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container

# **Implementation using Brute Force**

```python
# Brute Force
class PoorSolution(object):
    def computeArea(self, val1, val2, index_dis):
        val = min(val1, val2)
        return val * index_dis

    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        max_area = 0
        for index, num in enumerate(height):
            for index_next, num_next in enumerate(height[index + 1:], start=index + 1):
                area = self.computeArea(num, num_next, index_next - index)
                max_area = max(max_area, area)
        return max_area
```

```python
test_list = [1,8,6,2,5,4,8,3,7]
poor = PoorSolution()
poorrs = poor.maxArea(test_list)
print(poorrs)
```

# **Implementation using Two-Pointer Algorithm**

```python
class Solution(object):
    # def min(self, val1, val2):
    #     min = 0
    #     if val1 >= val2:
    #         min = val2
    #     else:
    #         min = val1
    #     return min
    # def max(self, val1, val2):
    #     max = 0
    #     if val1 >= val2:
    #         max = val1
    #     else:
    #         max = val2
    #     return max
    def maxArea(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        max_area = 0
        left = 0
        right = len(height) - 1
        while left < right:
            current_area = (right - left) * min(height[right], height[left])
            max_area = max(current_area, max_area)
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
```

```python
# Test
test_list = [1,8,6,2,5,4,8,3,7]
rs = Solution().maxArea(test_list)
print(rs)
```

# **Review**

Using Brute Force, the time complexity is O(n^2) (having iters in iters).
Using Two-Pointer Algorithm, the time complexity is becoming O(n) due to creating a window in the largest view and narrow down the window gradually arcording to the value of margin of left and right.

Because two approachs both use no more extra space, their space complexity is O(1).
