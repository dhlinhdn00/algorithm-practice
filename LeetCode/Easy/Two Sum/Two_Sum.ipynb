#**LeetCode: Two Sum**

# **Problem**


**Reference:**

https://leetcode.com/problems/two-sum/description/

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.



***Example 1:***

Input: nums = [2,7,11,15], target = 9

Output: [0,1]

Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

***Example 2:***

Input: nums = [3,2,4], target = 6

Output: [1,2]

***Example 3:***

Input: nums = [3,3], target = 6

Output: [0,1]

***Constraints:***

2 <= nums.length <= 104

-109 <= nums[i] <= 109

-109 <= target <= 109

Only one valid answer exists.


**Follow-up:** Can you come up with an algorithm that is less than O(n2) time complexity?

# **Code**

```python
# My code
# Brute Force with little improvement
# the worst case: O[n^2]
class MySolution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        lenght = len(nums)
        for i in range(lenght):
            rest = target - nums[i]
            if rest in nums[i+1:]:
                return [i, nums[i+1:].index(rest)+i+1]
        return []
```

```python
nums = [1, 4, 6, 6, 8]
target = 10
print(MySolution.twoSum(MySolution(), nums, target))
```

```python
# The most optimal
# Hash Table
# the worst case: O[n]
class BestSolution:
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        hash_table = {}

        for i, num in enumerate(nums):
            rest = target - num

            if rest in hash_table:

                return [hash_table[rest], i]

            # print(f"Trước {hash_table}")
            hash_table[num] = i
            # print(f"Sau {hash_table}")
        return []
```

```python
nums = [1, 4, 5, 3, 8, 6]
target = 10
print(BestSolution.twoSum(BestSolution(), nums, target))
```

# **Review**

Two approaches of mysolution and bestsolution are the same, but the data structures that are used to store and access the data are different.

The different comes from how hash table and array work:
Linear for Array and Almost constant time for hash table.
