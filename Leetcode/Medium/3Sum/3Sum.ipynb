#**LeetCode: 3Sum**

# **Problem**


**Reference:**

https://leetcode.com/problems/3sum/description/

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

# **Implementation**

```python
class Solution(object):
    @staticmethod
    def quicksort(arr):
        if len(arr) <= 1:
            return arr
        else:
            pivot = arr[len(arr) // 2]
            left = [x for x in arr if x < pivot]
            middle = [x for x in arr if x == pivot]
            right = [x for x in arr if x > pivot]
            return Solution.quicksort(left) + middle + Solution.quicksort(right)

    def threeSum(self, arr):
        """
        :type arr: List[int]
        :rtype: List[List[int]]
        """
        sorted_arr = Solution.quicksort(arr)
        result = []
        for i in range(len(sorted_arr) - 2):
            if i > 0 and sorted_arr[i] == sorted_arr[i-1]:
                continue
            left = i + 1
            right = len(sorted_arr) - 1
            while left < right:
                current_sum = sorted_arr[i] + sorted_arr[left] + sorted_arr[right]
                if current_sum == 0:
                    result.append([sorted_arr[i], sorted_arr[left], sorted_arr[right]])
                    left += 1
                    right -= 1
                    while left < right and sorted_arr[left] == sorted_arr[left - 1]:
                        left += 1
                    while left < right and sorted_arr[right] == sorted_arr[right + 1]:
                        right -= 1
                elif current_sum < 0:
                    left += 1
                else:
                    right -= 1
        return result
```

```python
arr = [-1,0,1,2,-1,-4]
# Solution.quicksort(arr)
Solution().threeSum(arr)
```
