#**LeetCode: Longest Substring Without Repeating Characters**

# **Problem**


**Reference:**

https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

Given a string s, find the length of the longest
substring without repeating characters.



***Example 1:***

Input: s = "abcabcbb"
Output: 3

Explanation: The answer is "abc", with the length of 3.

***Example 2:***

Input: s = "bbbbb"

Output: 1

Explanation: The answer is "b", with the length of 1.

***Example 3:***

Input: s = "pwwkew"
Output: 3

Explanation: The answer is "wke", with the length of 3.

Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.


***Constraints:***

0 <= s.length <= 5 * 104

s consists of English letters, digits, symbols and spaces.

# **My try**

```python
# Brute Force
class MySolution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        max_len = 0
        for i in range(len(s)):
          for j in range(len(s)+1):
            if len(s[i:j]) == len(set(s[i:j])):
              # print(s[i:j])
              # print(max_len)
              if len(s[i:j]) > max_len:
                max_len = len(s[i:j])

        return max_len
```

# **My submission**

```python
# GPT 's Solution
# Window Sliding
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        char_map = {}
        left = 0
        max_length = 0

        for right in range(len(s)):
            if s[right] in char_map:
                # print("Debug")
                # print("s[right] = ", s[right])
                # print("char_map[s[right]] = ", char_map[s[right]])
                left = max(left, char_map[s[right]] + 1)
                print("left = ", left)
            char_map[s[right]] = right
            max_length = max(max_length, right - left + 1)

        return max_length
```

```python
example1 = "abcabcbb"
print(Solution().lengthOfLongestSubstring(example1))  # Kết quả: 3
```

```python
example2 = "bbbbb"
print(Solution().lengthOfLongestSubstring(example2))  # Kết quả: 1
```

```python
example3 = "pwwkew"
print(Solution().lengthOfLongestSubstring(example3))  # Kết quả: 3
```
