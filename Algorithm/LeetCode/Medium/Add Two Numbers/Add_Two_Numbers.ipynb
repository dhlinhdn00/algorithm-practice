You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
class LinkedList:
  def __init__(self):
    self.head = None
  def append(self, data):
    "Add a new node at the end of the list"
    if not self.head:
      self.head = Node(data)
    else:
      current = self.head
      while current.next:
        current = current.next
      current.next = Node(data)
  def prepend(self, data):
    "Add a new node at the first of the list"
    new_head = Node(data)
    new_head.next = self.head
    self.head = new_head
  def print_list(self):
    current = self.head
    while current:
      print(current.data, end=" ")
      current = current.next
    print()
  @classmethod
  def from_list(cls, data_list):
      """
      Creates a new instance of the LinkedList from a provided list of elements.

      This method is a class method, which is chosen over an instance method or
      an external function for several reasons:

      1. **Class Scope:** As a class method, it operates on the class itself rather
        than an instance of the class. This is ideal for factory methods like this
        one, where the goal is to create new instances of the class from a given
        dataset. Using `cls` allows this method to be inherited and potentially
        overridden by subclasses, offering flexibility in extending the class's
        functionality.

      2. **No Need for an Existing Instance:** Unlike instance methods, which require
        an object of the class to be already created in order to be called, class
        methods can be invoked directly on the class. This makes `from_list` convenient
        for creating new LinkedList instances from scratch without first needing an
        empty LinkedList object.

      3. **Not Suitable for External Functions:** While it's possible to implement
        this functionality as an external function, encapsulating it within the
        LinkedList class as a class method keeps the code organized and cohesive.
        It logically groups the method with the class it operates on, making the
        class self-contained and easier to understand and maintain.

      This method uses `cls()` to create a new LinkedList instance, ensuring that
      if the method is called on a subclass, it correctly returns an instance of
      that subclass, preserving the expected object-oriented behaviors.

      Args:
      data_list (list): A list of data elements to be added to the new LinkedList.

      Returns:
      LinkedList: A new instance of the LinkedList (or subclass thereof) containing
                  the elements from `data_list`.
      """
      # ll = LinkedList()
      ll = cls()  # Create a new LinkedList instance using cls for subclass compatibility and inheritance
      for data in data_list:
          ll.append(data)
      return ll
  @classmethod
  def from_num_reserve(cls, num):
    ll = cls()
    for char in str(num):
      ll.prepend(int(char))
    return ll
  @classmethod
  def from_num(cls, num):
    ll = cls()
    for char in str(num):
      ll.append(int(char))
    return ll
  def to_list(self):
    list_ouput = []
    current = self.head
    while current:
      list_ouput.append(current.data)
      current = current.next
    return list_ouput
  def to_num_reserve(self):
    num_str = ''
    current = self.head
    while current:
      num_str = str(current.data) + num_str  # Prepends the digit
      current = current.next
    return int(num_str) if num_str else 0
```

```python
# My Solution
class ListNode(object):
  def __init__(self, val=0, next=None):
    self.val = val
    self.next = next

  @staticmethod
  def from_list(data_list):
    if not data_list:
      return None

    head = ListNode(data_list[0])
    current = head
    for data in data_list[1:]:
      current.next = ListNode(data)
      current = current.next
    return head

  @staticmethod
  def to_list(head):
    result = []
    current = head
    while current:
        result.append(current.val)
        current = current.next
    return result

  @staticmethod
  def from_number_reserve(number):
    head = None
    for digit in str(number):
        if head:
            new_node = ListNode(int(digit), head)
            head = new_node
        else:
            head = ListNode(int(digit))
    return head or ListNode(0)

  @staticmethod
  def to_number_reserve(head):
    number = 0
    multiplier = 1
    current = head
    while current:
        number += current.val * multiplier
        multiplier *= 10
        current = current.next
    return number
```

```python
l1 = ListNode.from_number_reserve(12345)
print(ListNode.to_list(l1))
ListNode.to_number_reserve(l1)
```

```python
class MySolution(object):
    def addTwoNumbers(self, l1, l2):
        n1 = ListNode.to_number_reserve(l1)
        n2 = ListNode.to_number_reserve(l2)
        n3 = n1 + n2
        l3 = ListNode.from_number_reserve(n3)
        print(ListNode.to_list(l3))
        return l3
```

# **ChatGPT Solution**

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        temp = ListNode(0)
        current = temp
        carry = 0

        while l1 or l2 or carry:
            val1 = (l1.val if l1 else 0)
            val2 = (l2.val if l2 else 0)
            sum = val1 + val2 + carry
            carry = sum // 10
            sum = sum % 10

            current.next = ListNode(sum)
            current = current.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return temp.next

def toListNode(numbers):
    dummy_root = ListNode(0)
    ptr = dummy_root
    for number in numbers:
        ptr.next = ListNode(number)
        ptr = ptr.next
    return dummy_root.next

def printListNode(node):
    while node:
        print(node.val, end=' ')
        node = node.next
    print()
```

```python
# Test cases
sol = Solution()

# Test case 1
l1 = toListNode([2, 4, 3])  # Represents 342
l2 = toListNode([5, 6, 4])  # Represents 465
result = sol.addTwoNumbers(l1, l2)
printListNode(result)  # Expected output: 7 0 8, which represents 807

# Test case 2
l1 = toListNode([0])  # Represents 0
l2 = toListNode([0])  # Represents 0
result = sol.addTwoNumbers(l1, l2)
printListNode(result)  # Expected output: 0

# Test case 3
l1 = toListNode([9,9,9,9,9,9,9])  # Represents 9999999
l2 = toListNode([9,9,9,9])  # Represents 9999
result = sol.addTwoNumbers(l1, l2)
printListNode(result)  # Expected output: 8 9 9 9 0 0 0 1, which represents 10009998
```
